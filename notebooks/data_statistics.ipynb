```python
stats = {}

for p in data_path.iterdir():
    stats[p.name] = {}
    ds_dict = load_from_disk(p)
    names = ds_dict[split].features["labels"].names

    s = {}
    for split in ds_dict:

        df = ds_dict[split].to_pandas()
        
        df = pd.concat(
            [df["labels"].value_counts(normalize=True), df["labels"].value_counts()],
            axis=1,
            keys=("perc", "count"),
        )
        df["labels"] = [names[i] for i in df.index]
        df = df.sort_index(ascending=True)[df.columns.tolist()[::-1]]
        s[split] = pd.concat([df, df.sum(0).to_frame().T])
    
    df = pd.concat(s)
    # df.index = df.index.get_level_values(0)
    df.columns = ["Label", "Count", "Proportion"]
    stats[p.name] = df
```
